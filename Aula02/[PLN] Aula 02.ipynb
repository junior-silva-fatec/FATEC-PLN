## Exemplo 02 - Atribuição de variáveis

```python
# Em Python, tudo é tratado como objeto
carros = 100
  # objeto = numero inteiro, valor = 100 - rótulo ---> nome da variável [carros]
espaco_no_carro = 4.0 # numero real ou ponto flutuante
motoristas = 30
passageiros = 90
carros_sem_motoristas = carros - motoristas
carros_com_motoristas = motoristas
capacidade_de_passageiros = carros_com_motoristas * espaco_no_carro
media_passageiros_por_carro = passageiros / carros_com_motoristas

print("Existem", carros, "disponíveis.")
print(f"Existem apenas {motoristas} carros disponíveis.")
print("Nós podemos transportar {} pessoas hoje.".format(capacidade_de_passageiros))
print(100/5)
print("{} + {} + {} + {}".format(media_passageiros_por_carro, capacidade_de_passageiros, carros_com_motoristas, espaco_no_carro))
print(f"{media_passageiros_por_carro} * {capacidade_de_passageiros} * {carros_com_motoristas} * {espaco_no_carro}")
  # as operações moldam o tipo de dado que vai ser apresentado
  # int / int = real
```

## Exemplo 06 - Laços de repetição
while   

for contador in range(contagem):  
    instruções a serem repetidas

```python
i = 0

# variável simples: armazena apenas um valor
# variável composta: armazena vários valores
  # listas: são mutáveis. Ex.: [a, b, c, 45, a, b, [ad]]
  # tuplas: imutáveis. Ex.: (a, b, c, (asd, qwer), a, c)
  # conjuntos: mutáveis, mas sem repetição. Ex.: {valor1, valor2, valor3, ...}

numero = [] # inicializar uma variável composta do tipo lista

while i < 6:
  print(f"Estou indo de 0 até {i}")
  numero.append(i)
  i = i + 1
  print(f'o valor da lista é {numero}')
  # aqui estou dentro da repetição

# aqui estou fora da repetição

print(f"\n\nLaço for\n")
for i in range(0, 6):
  print(f"Estou indo de 0 até {i}")
  numero.append(i)
  print(f"o valor da lista é {numero}")

print(f"\n\nLaço for com strings\n")
texto = "Aula de sexta-feira"

for i in texto:
  print(f"O contador está com a letra {i}")
```

```python
# Exercício 01
print("EXERCÍCIO 01 - CÁLCULO DO VOLUME DE UMA CAIXA RETANGULAR\n")
comprimento = float(input("Digite o comprimento da caixa: "))
largura = float(input("Digite a largura da caixa: "))
altura = float(input("Digite a altura da caixa: "))
volume = comprimento * altura * largura
print(f"\nO volume da caixa é {volume} unidades.")
```

```python
# Exercício 02
print("EXERCÍCIO 02 - PROGRAMA QUE LÊ UM TEXTO E CONTA O NÚMERO DE PALAVRAS\n")
texto = input("Digite o texto: ")
palavras = texto.split()
print(f"O texto contém {len(palavras)} palavras.")
```

```python
# Exercício 03
print("EXERCÍCIO 03 - BUSCAR E SUBSTITUIR PALAVRAS EM UM TEXTO\n")
texto = input("Digite o texto: ")
palavra_antiga = input("Digite a palavra a ser substituída: ")
palavra_nova = input("Digite a nova palavra: ")
novo_texto = texto.replace(palavra_antiga, palavra_nova)
print("\nO novo texto é:")
print(novo_texto)
```

```python
# Exercício 04
print("EXERCÍCIO 04 - SEPARAR FRASES USANDO SPLIT() E ITERAR SOBRE ELAS\n")
import re
texto = input("Digite o texto: ")
frases = re.split(r'[.?!]', texto)
for frase in frases:
  frase = frase.strip()
  if frase:
    print(frase)
```

```python
# Exercício 05
print("EXERCÍCIO 05 - LER TRÊS VALORES INTEIROS REPRESENTADOS PELAS VARIÁVEIS A, B, E C \nE APRESENTAR OS VALORES LIDOS DISPOSTOS EM ORDEM CRESCENTE\n")
a = int(input("Digite o valor da variável a: "))
b = int(input("Digite o valor da variável b: "))
c = int(input("Digite o valor da variável c: "))
valores = [a, b, c]
valores_ordem_crescente = sorted(valores)
print("\nValores a, b, e c em ordem crescente: ", valores_ordem_crescente)
```

```python
# Exercício 06
print("EXERCÍCIO 06 - FAZER A LEITURA DE QUATRO VALORES NUMÉRICOS INTEIROS REPRESENTADOS \nPELAS VARIÁVEIS A, B, C E D. APRESENTAR APENAS OS VALORES QUE SEJAM DIVISÍVEIS \nPOR 2 E 3.\n")
a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))
d = int(input("Digite o valor de d: "))
valores = [a, b, c, d]
numeros_divisiveis = [valor for valor in valores if valor % 2 == 0 and valor % 3 == 0]
if numeros_divisiveis:
  print("Números divisíveis por 2 e 3:", numeros_divisiveis)
else:
  print("Nenhum dos números informados é divisível por 2 e 3")
```

```python
# Exercício 07
print("EXERCÍCIO 07 - CONSTRUIR UM PROGRAMA QUE APRESENTE TODOS OS VALORES NUMÉRICOS \nDIVISÍVEIS POR 4 E MENORES QUE 200.\n")

for numero in range(4, 200, 4):
  print(numero)
```
